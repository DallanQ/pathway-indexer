# Get the links from the origin


```python
import pandas as pd
```

```python
filename = "../data/qa_list/index_questions.csv"

df = pd.read_csv(
    filename,
    header=1,
)
df.index = range(1, len(df) + 1)
df.head()
```

```python
df.dropna(subset=["Link to Ideal Answer"], inplace=True)
df.head(5)
```

```python
df_no_links = df[df["Link to Ideal Answer"].str.contains("http") == False]

# make a drop of the rows that have no link to ideal answer
df.drop(df_no_links.index, inplace=True)
df.head(10)
```

```python
df.iloc[7]["Link to Ideal Answer"]
```

```python
# Función para limpiar y extraer los links
def extract_links(text):
    # Dividir el texto por los saltos de línea y filtrar solo los enlaces
    links = [
        line.strip() for line in text.split("\n") if line.strip().startswith("https")
    ]
    return links
```

```python
# some of the rows has multiple links, we will convert them to a list
df["Link to Ideal Answer"] = df["Link to Ideal Answer"].apply(extract_links)
df.head(10)

# clean_text = text.split("https", 1)[-1]
# clean_text = "https" + clean_text
```

```python
df.iloc[4]["Link to Ideal Answer"]
```

```python
list_of_links = df["Link to Ideal Answer"].tolist()

# flat the list of lists
flat_list = [item for sublist in list_of_links for item in sublist]
```

```python
print("len:", len(flat_list))
```

# Verify inexistent links


```python
filename = "output_data.csv"

df_exist = pd.read_csv(
    filename,
    header=0,
)

df_exist.head()
```

```python
existing = df_exist["URL"].tolist()
```

```python
# Validate if the link is already in the existing list

new_links = [link for link in flat_list if link not in existing]
```

```python
len(new_links)

# Save the links in a txt file named new_links.txt
with open("../data/temporary/new_links.txt", "w") as file:
    for link in new_links:
        file.write(link + "\n")
```
