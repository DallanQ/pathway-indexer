```python
import os
import pandas as pd
```

```python
# load links from data_3_11
all_links_filename = "../data/data_3_11/all_links.csv"


df_links = pd.read_csv(all_links_filename)
df_links.head()
```

```python
# filter only the URL with #
df_links = df_links[df_links["URL"].str.contains("#")]
# sort by URL
df_links = df_links.sort_values(by=["URL"])
df_links
```

```python
# filter before the first # and show all those URLs, sorted
df_links["URL2"] = df_links["URL"].apply(lambda x: x.split("#")[0])
# show if the URL2 appears more than once
df_links = df_links[df_links.duplicated(subset=["URL2"], keep=False)]
df_links = df_links.sort_values(by=["URL2"])
# drop URL2
df_links = df_links.drop(columns=["URL2"])
df_links
```

```python
df_links["URL"] = df_links["URL"].str.split("#").str[0]
df_links
```
