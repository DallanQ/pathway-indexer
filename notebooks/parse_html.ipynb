```python
import os
from bs4 import BeautifulSoup
from markdownify import markdownify as md
```

```python

html_dir = '../data/html'
markdown_dir = '../data/markdown'

# mkdir md if not exists
os.makedirs(markdown_dir, exist_ok=True)


def html_to_markdown(html_content):
    '''Takes html and converts the main components to markdown'''
    soup = BeautifulSoup(html_content, 'html.parser')

    # Decompose unwanted tags
    for tag in soup.find_all(['nav', 'header', 'footer', 'aside', 'img', 'svg', 'style', 'script']):
        tag.decompose()

    # Get the main content
    main_content = soup.find('main') or soup

    # Convert the remaining HTML to Markdown, keeping specific tags
    markdown_content = md(str(main_content), strip=['img', 'svg', 'style', 'script', 'nav', 'header', 'footer', 'aside'])

    return markdown_content

# Directories
html_dir = '../data/html'  # Replace with your HTML directory
markdown_dir = '../data/markdown'  # Replace with your Markdown directory

# Ensure markdown_dir exists
os.makedirs(markdown_dir, exist_ok=True)

# Handle each HTML file
for filename in os.listdir(html_dir):
    if filename.endswith('.html'):
        html_path = os.path.join(html_dir, filename)
        with open(html_path, 'r', encoding='utf-8') as html_file:
            html_content = html_file.read()

        # Convert HTML to Markdown
        markdown_content = html_to_markdown(html_content)

        # Define markdown file path
        markdown_filename = filename.replace('.html', '.md')
        markdown_path = os.path.join(markdown_dir, markdown_filename)
        
        # Check if the markdown file already exists
        if os.path.exists(markdown_path):
            print(f"Markdown file {markdown_filename} already exists. Skipping.")
        else:
            # Save Markdown file
            with open(markdown_path, 'w', encoding='utf-8') as markdown_file:
                markdown_file.write(markdown_content)

            print(f"Converted {filename} to {markdown_filename}")
```
